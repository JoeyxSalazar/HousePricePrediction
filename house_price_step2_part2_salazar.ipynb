```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import seaborn as sn
sn.set(style='whitegrid')
```

Load training Data

```python
x_train = pd.read_csv('housing_train_clean.csv')
y_train = x_train['SalePrice']
x_train.drop('SalePrice', axis = 1)
```

```python
y_train
```

Load Test Data

```python
x_test = pd.read_csv('housing_test_clean.csv')
y_test = x_test['SalePrice']
x_test.drop('SalePrice', axis = 1)
```

```python
y_test
```

Convert Pandas to Numpy Array

```python
x_train = x_train.values
y_train = y_train.values.reshape(-1)
x_test = x_test.values
y_test = y_test.values.reshape(-1)
```

Normalize x_train and x_test using minmaxscaler

```python
scaler=MinMaxScaler()
#get the parameters of the transform
scaler.fit(x_train)
#normalize the features in the training set
x_train_s = scaler.transform(x_train)
#normalize the features in the test set
x_test_s = scaler.transform(x_test)

from sklearn.model_selection import train_test_split

x_train_s, x_val_s, y_train, y_val = train_test_split(x_train_s, y_train, test_size=0.1, random_state=0)
```

KNeighborsRegressor (n_neighbors)

parameter range: 1 to 100, with the step of 10

```python
from sklearn.neighbors import KNeighborsRegressor
MAPEsKNN = {}
for x in range(1, 101, 10):
    knn = KNeighborsRegressor(n_neighbors = x)
    # train the KNN regressor
    knn.fit(x_train_s, y_train)
    # test the KNN regressor
    Y_val_pred_knn = knn.predict(x_val_s)
    # Calculate errors on test set
    #MSE = np.mean((y_test - Y_test_pred_knn)**2)
    #MAE = np.mean(np.abs(y_test - Y_test_pred_knn))
    MAPE =  np.mean(np.abs(y_val - Y_val_pred_knn)/y_val)
    print(x, ':     ', MAPE)
    MAPEsKNN[x] = MAPE
print()
print('Optimal n_neighbors : MAPE')
print(min(MAPEsKNN, key = MAPEsKNN.get), end=': ')
print(min(MAPEsKNN.values()))
```

```python
knn = KNeighborsRegressor(n_neighbors = 11)
# train the KNN regressor
knn.fit(x_train_s, y_train)
# test the KNN regressor
Y_test_pred_knn = knn.predict(x_test_s)
# Calculate errors on test set
MSE1 = np.mean((y_test - Y_test_pred_knn)**2)
MAE1 = np.mean(np.abs(y_test - Y_test_pred_knn))
MAPE1 =  np.mean(np.abs(y_test - Y_test_pred_knn)/y_test)
```

DecisionTreeRegressor

```python
from sklearn.tree import DecisionTreeRegressor
MAPEsDT = {}
for y in range(1, 101, 10):
    model = DecisionTreeRegressor(max_depth = y)
    model.fit(x_train_s, y_train)
    # test the regressor
    Y_val_pred_dt = model.predict(x_val_s)
    # Calculate errors
    #MSE3 = np.mean((y_test - Y_test_pred_dt)**2)
    #MAE3 = np.mean(np.abs(y_test - Y_test_pred_dt))
    MAPE =  np.mean(np.abs(y_val - Y_val_pred_dt)/y_val)
    print(y, ':     ', MAPE)
    MAPEsDT[y] = MAPE
print()
print('Optimal max_depth : MAPE')
print(min(MAPEsDT, key = MAPEsDT.get), end=': ')
print(min(MAPEsDT.values()))

```

```python
model = DecisionTreeRegressor(max_depth = 11)
model.fit(x_train_s, y_train)
# test the regressor
Y_test_pred_dt = model.predict(x_test_s)
# Calculate errors
MSE2 = np.mean((y_test - Y_test_pred_dt)**2)
MAE2 = np.mean(np.abs(y_test - Y_test_pred_dt))
MAPE2 =  np.mean(np.abs(y_test - Y_test_pred_dt)/y_test)
```

RandomForestRegressor

```python
from sklearn.ensemble import RandomForestRegressor
MAPEsRF = {}
for z in range(1, 101, 10):
    RF = RandomForestRegressor(max_depth = z)
    RF.fit(x_train_s, y_train)
    Y_val_pred_rf = RF.predict(x_val_s)
    MAPE =  np.mean(np.abs(y_val - Y_val_pred_rf)/y_val)
    print(z, ':     ', MAPE)
    MAPEsRF[z] = MAPE
print()
print('Optimal max_depth : MAPE')
print(min(MAPEsRF, key = MAPEsRF.get), end=': ')
print(min(MAPEsRF.values()))
    
```

```python
model = DecisionTreeRegressor(max_depth = 51)
model.fit(x_train_s, y_train)
# test the regressor
Y_test_pred_dt = model.predict(x_test_s)
# Calculate errors
MSE3 = np.mean((y_test - Y_test_pred_dt)**2)
MAE3 = np.mean(np.abs(y_test - Y_test_pred_dt))
MAPE3 =  np.mean(np.abs(y_test - Y_test_pred_dt)/y_test)
```

Construct Dataframe

```python
dict = {
        'KNN':[MSE1, MAE1, MAPE1],
        'Decision Tree':[MSE2, MAE2, MAPE2],
        'Random Forest':[MSE3, MAE3, MAPE3]}

results = pd.DataFrame(dict, index = ['MSE' ,'MAE', 'MAPE' ])
results
```
