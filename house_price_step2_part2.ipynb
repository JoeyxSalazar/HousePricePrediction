```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
import seaborn as sn
sn.set(style='whitegrid')
```

```python
X_train = pd.read_csv("housing_train_clean.csv")

X_train =  X_train.drop('SalePrice', axis=1)
X_train.head()
```

```python
data_train = pd.read_csv("housing_train.csv")

Y_train = data_train['SalePrice']
Y_train.head()
```

```python
X_test = pd.read_csv("housing_test_clean.csv")
X_test = X_test.drop('SalePrice', axis=1)
X_test
```

```python
Y_test = pd.read_csv('housing_test_y.csv')
Y_test = Y_test['SalePrice']
Y_test
```

```python
X_train=X_train.values
Y_train=Y_train.values.reshape(-1)
X_test=X_test.values
Y_test=Y_test.values.reshape(-1)
```

```python
print(X_test.shape, Y_test.shape)
```

```python
#Normalize features

scaler=MinMaxScaler()
#get the parameters of the transform
scaler.fit(X_train)
#normalize the features in the training set
X_train_s = scaler.transform(X_train)
#normalize the features in the test set
X_test_s = scaler.transform(X_test)

from sklearn.model_selection import train_test_split

X_train_s, X_val_s, Y_train, Y_val = train_test_split(X_train_s, Y_train, test_size=0.1, random_state=0)
```

```python
# KNeighborsRegressor (n_neighbors)
# DecisionTreeRegressor (max_depth)
# RandomForestRegressor (max_depth)
# (see examples presented in class)
# parameter range: 1 to 100, with the step of 10
```

```python
# K nieghbors regression 
from sklearn.neighbors import KNeighborsRegressor
MSE_K_list = []
MAE_K_list = []
MAPE_K_list = []
for x in range(1, 101, 10):
    knn = KNeighborsRegressor(n_neighbors = x)
    # train the KNN regressor
    knn.fit(X_train_s, Y_train)
    # test the KNN regressor
    Y_val_pred = knn.predict(X_val_s)
    # Calculate errors on validation set
    MSE_K = np.mean((Y_val - Y_val_pred)**2)
    MAE_K = np.mean(np.abs(Y_val - Y_val_pred))
    MAPE_K =  np.mean(np.abs(Y_val - Y_val_pred)/Y_val)
    MSE_K_list.append(MSE_K)
    MAE_K_list.append(MAE_K)
    MAPE_K_list.append(MAPE_K)
#     print('MSE=', MSE_K)
#     print('MAE=', MAE_K)
    print('MAPE=', MAPE_K, 'for', x)

# min_MSE_K = min(MSE_K_list)
# min_MAE_K = min(MAE_K_list)
# min_MAPE_K = min(MAPE_K_list)

# print("min mse:", min_MSE_K)
# print("min mae:", min_MAE_K)
# print("min mape:", min_MAPE_K)
```

```python
# testing the K neighbors Regressor on the test set

knn = KNeighborsRegressor(n_neighbors = 11)
# train the KNN regressor
knn.fit(X_train_s, Y_train)
# test the KNN regressor
Y_test_pred = knn.predict(X_test_s)
# Calculate errors on validation set
MSE_K = np.mean((Y_test - Y_test_pred)**2)
MAE_K = np.mean(np.abs(Y_test - Y_test_pred))
MAPE_K =  np.mean(np.abs(Y_test - Y_test_pred)/Y_test)
# MSE_K_list.append(MSE_K)
# MAE_K_list.append(MAE_K)
# MAPE_K_list.append(MAPE_K)
print('MSE=', MSE_K)
print('MAE=', MAE_K)
print('MAPE=', MAPE_K)
```

```python
# DecisionTreeRegressor
from sklearn.tree import DecisionTreeRegressor
```

```python

MSE_T_list = []
MAE_T_list= []
MAPE_T_list = []

for y in range(1, 101, 10):
    model = DecisionTreeRegressor(max_depth=y)
    model.fit(X_train_s, Y_train)

    # test the regressor
    Y_val_pred = model.predict(X_val_s)
    # Calculate errors
    MSE_T = np.mean((Y_val - Y_val_pred)**2)
    MAE_T = np.mean(np.abs(Y_val - Y_val_pred))
    MAPE_T =  np.mean(np.abs(Y_val - Y_val_pred)/Y_val)
#     print('MSE=', MSE_T)
#     print('MAE=', MAE_T)
    print('MAPE=', MAPE_T, 'for', y)
    MSE_T_list.append(MSE_T)
    MAE_T_list.append(MAE_T)
    MAPE_T_list.append(MAPE_T)

# min_MSE_T = min(MSE_T_list)
# min_MAE_T =  min(MAE_T_list)
# min_MAPE_T = min(MAPE_T_list)

# print("min mse:", min_MSE_T)
# print("min mae:", min_MAE_T)
# print("min mape:", min_MAPE_T)
    
```

```python
# best MAPE= 0.132044151366305 for 11
model = DecisionTreeRegressor(max_depth=11)
model.fit(X_train_s, Y_train)

# test the regressor
Y_test_pred = model.predict(X_test_s)
# Calculate errors
MSE_T = np.mean((Y_test - Y_test_pred)**2)
MAE_T = np.mean(np.abs(Y_test - Y_test_pred))
MAPE_T =  np.mean(np.abs(Y_test - Y_test_pred)/Y_test)
#     print('MSE=', MSE_T)
#     print('MAE=', MAE_T)
print('MAPE=', MAPE_T)
MSE_T_list.append(MSE_T)
MAE_T_list.append(MAE_T)
MAPE_T_list.append(MAPE_T)
```
